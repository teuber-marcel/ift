```python
import pyift.pyift as ift
import matplotlib.pyplot as plt
```

```python
img = ift.ReadImageByExt("images/dog.png")
A   = ift.Circular(1.0)
```

```python
seeds = ift.ReadSeeds(img, "seeds/dog-seeds.txt")

basins = ift.ImageBasins(img, A)

label = ift.Watershed(basins, A, seeds, None)
```

```python
plt.imshow(img.ToPlot())
```

```python
plt.imshow(label.ToPlot())
```

```python
border = ift.BorderImage(label,False);
A      = ift.Circular(1.0);
RGB    = ift.Color()
RGB.SetValues((255,0,0))
YCbCr  = ift.RGBtoYCbCr(RGB,255)
ift.DrawBorders(img,border,A,YCbCr,A)
ift.DrawSeeds(img,seeds)
plt.imshow(img.ToPlot())
```
