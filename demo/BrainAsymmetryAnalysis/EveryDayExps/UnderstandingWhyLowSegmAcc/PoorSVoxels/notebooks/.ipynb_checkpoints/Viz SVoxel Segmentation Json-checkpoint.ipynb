```python
import os
from os.path import join as pjoin
import pdb
import json

import numpy as np

import plotly.graph_objs as go
import plotly as py
import ipywidgets as widgets
py.offline.init_notebook_mode(connected=True)
```

```python
os.system(f"make iftExtractSVoxelStats -C $NEWIFT_DIR/demo/BrainAsymmetryAnalysis/EveryDayExps/UnderstandingWhyLowSegmAcc/PoorSVoxels")
```

```python
work_dir = "/Users/hisamuka/workspace/exps/UnderstandingWhyLowSegmentation"
saad_result_dir = os.path.join(work_dir, "exps/SAAD")
gt_dir = os.path.join(work_dir, "bases/ATLAS-304/3T/regs/nonrigid/labels/primary_stroke")
```

```python
brain_mask_path = os.path.join(work_dir, "template/mni152_nonlinear_sym_brain_envelope.nii.gz")
svoxels_stats_json = os.path.join(work_dir, "exps/svoxels_stats.json")

os.system(f"iftExtractSVoxelStats --saad-result-dir {saad_result_dir} --ground-truth-dir {gt_dir} --brain-envelope-mask {brain_mask_path} --output-json {svoxels_stats_json}")

stats = {}
with open(svoxels_stats_json, 'r') as f:
        stats = json.load(f)

print(stats)
```

```python
stats_info = stats["detected"]
min_svoxel_volume = min(stats_info["svoxels-volume"])
max_svoxel_volume = max(stats_info["svoxels-volume"])

data = [
    go.Parcoords(
        line = dict(color = '#33a067'),
        dimensions = list([
            dict(
                label = 'SVoxel Volume',
                values = stats_info["svoxels-volume"],
                tickvals = np.arange(min_svoxel_volume, max_svoxel_volume + 1, (max_svoxel_volume - min_svoxel_volume) / 20),
                tickformat = ".2f"),
            dict(
                label = 'Eucl. Distance of SVoxel Centroid to Brain Border',
                values = stats_info["edt"],
                tickformat = ".2f"),
            dict(
                label = '|SVoxel  GT| / |SVoxel| (Precision)',
                values = stats_info["precision"],
                range = [0.0, 1.0],
                tickformat = ".2f"),
#             dict(
#                 label = 'Perc. SVoxel out of GT (FDR)',
#                 values = fdr_dict[1],
#                 range = [0.0, 1.0],
#                 tickformat = ".2f"),
            dict(
                label = '|SVoxel  GT| / |GT| (TPR)',
                values = stats_info["tpr"],
                range = [0.0, 1.0],
                tickformat = ".2f"),
        ])
    )
]

layout = go.Layout(
    title = "Detected Supervoxels on GT"
)

fig = go.Figure(data = data, layout = layout)
py.offline.iplot(fig)
```

```python
stats_info = stats["missed"]
min_svoxel_volume = min(stats_info["svoxels-volume"])
max_svoxel_volume = max(stats_info["svoxels-volume"])

data = [
    go.Parcoords(
        line = dict(color = '#ee6e73'),
        dimensions = list([
            dict(
                label = 'SVoxel Volume',
                values = stats_info["svoxels-volume"],
                tickvals = np.arange(min_svoxel_volume, max_svoxel_volume + 1, (max_svoxel_volume - min_svoxel_volume) / 20),
                tickformat = ".2f"),
            dict(
                label = 'Eucl. Distance to Border',
                values = stats_info["edt"],
                tickformat = ".2f"),
            dict(
                label = '|SVoxel  GT| / |SVoxel| (Precision)',
                values = stats_info["precision"],
                range = [0.0, 1.0],
                tickformat = ".2f"),
#             dict(
#                 label = 'Perc. SVoxel out of GT (FDR)',
#                 values = fdr_dict[1],
#                 range = [0.0, 1.0],
#                 tickformat = ".2f"),
            dict(
                label = '|SVoxel  GT| / |GT| (TPR)',
                values = stats_info["tpr"],
                range = [0.0, 1.0],
                tickformat = ".2f"),
        ])
    )
]

layout = go.Layout(
    title = "Unetected Supervoxels on GT"
)

fig = go.Figure(data = data, layout = layout)
py.offline.iplot(fig)
```
