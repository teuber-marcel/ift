```python
import pyift.pyift as ift
import matplotlib.pyplot as plt
```

```python
# try images tools.png, bird1.png and bird2.png
bin = ift.ReadImageByExt("images/tools.png")
plt.imshow(bin.ToPlot())
```

```python
msskel = ift.ColorMSSkel2D(bin)
plt.imshow(msskel.ToPlot())
```

```python
A      = ift.Circular(1.5)
msskel = ift.MSSkel2D(bin, A, ift.IFT_INTERIOR, None, None)
msskel = msskel[0]
```

```python
skel   = ift.FThreshold(msskel,5.0,100.0,255);
term   = ift.TerminalPointSet2D(skel)
branch = ift.BranchPointSet2D(skel)
```

```python
border = ift.BorderImage(bin,False);
A      = ift.Circular(0.0);
RGB    = ift.Color()
RGB.SetValues((255,100,100))
YCbCr  = ift.RGBtoYCbCr(RGB,255)
img    = ift.CopyImage(skel)
ift.DrawBorders(img,border,A,YCbCr,A)
A      = ift.Circular(3.0);
RGB.SetValues((200,200,100))
YCbCr  = ift.RGBtoYCbCr(RGB,255)
ift.DrawPoints(img, term, YCbCr, A)
A      = ift.Circular(1.0);
RGB.SetValues((100,200,100))
YCbCr  = ift.RGBtoYCbCr(RGB,255)
ift.DrawPoints(img, branch, YCbCr, A)
plt.imshow(img.ToPlot())
```
